# Image filtering #

In this notebook we present various ways of image filtering problem.

Importing proper modules

```python
from PIL import Image, ImageFilter
from skimage import io, filters
import numpy as np
```

Loading the test image. Feel free to try your own images.

```python
im = Image.open("red_blood_cells.png")
im
```

```python
im_arr = np.array(im)
```

### Let's try PIL filters ###

```python
f1 = im.filter(ImageFilter.CONTOUR)
f1
```

```python
f2 = im.filter(ImageFilter.BLUR)
f2
```

```python
f3 = im.filter(ImageFilter.DETAIL)
f3
```

```python
f4 = im.filter(ImageFilter.EDGE_ENHANCE)
f4
```

```python
f5 = im.filter(ImageFilter.EDGE_ENHANCE_MORE)
f5
```

```python
f6 = im.filter(ImageFilter.EMBOSS)
f6
```

```python
f7 = im.filter(ImageFilter.SMOOTH)
f7
```

```python
f8 = im.filter(ImageFilter.SMOOTH_MORE)
f8
```

```python
f9 = im.filter(ImageFilter.SHARPEN)
f9
```

### Skimage filters###

```python
f10 = filters.sobel(im_arr[:,:,1])
io.imshow(f10)
io.show()
```

```python
f11 = filters.gaussian(im_arr[:,:,1])
io.imshow(f11)
io.show()
```

```python
f12 = filters.median(im_arr[:,:,1])
io.imshow(f12)
io.show()
```

```python
f13 = filters.scharr(im_arr[:,:,1])
io.imshow(f13)
io.show()
```

```python
f14 = filters.prewitt(im_arr[:,:,1])
io.imshow(f14)
io.show()
```

```python
f15 = filters.roberts(im_arr[:,:,1])
io.imshow(f15)
io.show()
```

```python
f16 = filters.laplace(im_arr[:,:,1])
io.imshow(f16)
io.show()
```

```python
f17 = filters.frangi(im_arr[:,:,1])
io.imshow(f17)
io.show()
```

```python
f18 = filters.hessian(im_arr[:,:,1])
io.imshow(f18)
io.show()
```
